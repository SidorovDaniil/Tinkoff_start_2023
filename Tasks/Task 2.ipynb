```python
# s = input()

s = 'thegorillaiswatching'
word = 'sheriff'
count_char = []
```

```python
def word_for_Jo(s, word):
    k = 0 

    for j in word:
        for i in s:
            if i == j:
                k = k + 1

        count_char.append(k)        
        k = 0
    
    count_char[-1] = round(count_char[-1]/2)
    count_char[-2] = round(count_char[-2]/2)
    
    return min(count_char)

print(word_for_Jo(s, word))
```
