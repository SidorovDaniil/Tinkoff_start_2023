```python
# a = input()
# b = input()
a = '5 900000'
b = '0 900001 0 0 0'
```

```python
a = list(map(int, list(a.split())))
b = list(map(int, list(b.split())))
prices = []
```

```python
a, b
```

```python
def price_for_Jo(a, b):
    for i in range(a[0]):
        if a[1] >= b[i]:
            final_price = b[i]
            prices.append(final_price)
        
    if len(prices) == 0:
        return 0
    else:
        return(max(prices))
    
print(price_for_Jo(a,b))
```
