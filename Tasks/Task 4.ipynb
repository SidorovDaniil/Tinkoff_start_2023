```python
n_m = input()
banknotes = input()

# n_m = '5 2'
# banknotes = '3 4'

n_m = list(map(int, n_m.split()))
banknotes = list(map(int, banknotes.split()))

all_banknotes = []

for i in range(n_m[1]):
    all_banknotes.append(banknotes[i])
    all_banknotes.append(banknotes[i])
    
all_banknotes = sorted(all_banknotes)
```

```python
def still_for_Jo(n_m, all_banknotes):
    for j in range(1, len(all_banknotes) + 1):
        need_to_still = n_m[0]
        still_banknotes = []
        k = 0
        
        for i in range(1, len(all_banknotes) + 1):
            if need_to_still - all_banknotes[-i] >= 0:
                need_to_still = need_to_still - all_banknotes[-i]
                still_banknotes.append(all_banknotes[-i])
                k = k + 1

        if need_to_still != 0:
            del all_banknotes[-1]
            
        if need_to_still == 0:
            return k, still_banknotes
        
        return False

if still_for_Jo(n_m, all_banknotes) == False:
    print(-1)
else:
    print(still_for_Jo(n_m, all_banknotes)[0])
    print(' '.join([str(x) for x in still_for_Jo(n_m, all_banknotes)[1]]))
```
