```python
number_of_card = int(input())
cards_jo = input()
cards_win = input()

# number_of_card = 5
# cards_jo = '1 4 2 2 4'
# cards_win = '1 4 4 2 2'

cards_jo = list(map(int, list(cards_jo.split())))
cards_win = list(map(int, list(cards_win.split())))
```

```python
def win_cond(number_of_card, cards_jo, cards_win):
    l = 0
    for i in range(number_of_card):
        if cards_jo[i] == cards_win[i]:
            l = l + 1

        else:
            break

    r = 0
    for i in range(1, number_of_card + 1):
        if cards_jo[-i] == cards_win[-i]:
            r = r + 1

        else:
            break
            
    cards_jo_shuffle = sorted(cards_jo[l:number_of_card-r])
    
    if cards_jo_shuffle == cards_win[l:number_of_card-r]:
        return 'YES'

    else:
        return 'NO'
    
    
print(win_cond(number_of_card, cards_jo, cards_win))
```
